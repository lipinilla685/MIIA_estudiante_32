```python
# código
#LAURA I. PINILLA MESA

lista_archivos = ["./Archivos/edad.txt", "./Archivos/escolaridad.txt", "./Archivos/estado_civil.txt", "./Archivos/estrato.txt", "./Archivos/genero.txt", "./Archivos/promedio.txt", "./Archivos/region.txt"]

def archivo_a_lista(directorio):
        f = open(directorio, mode="r", encoding="utf-8")
        lista_directorio = f.read().splitlines()
        f.close()
        return lista_directorio

for i in lista_archivos:
    if i==lista_archivos[0]:
        edad = archivo_a_lista(i)
    elif i == lista_archivos[1]:
        escolaridad = archivo_a_lista(i)
    elif i == lista_archivos[2]:
        estado_civil = archivo_a_lista(i)
    elif i == lista_archivos[3]:
        estrato = archivo_a_lista(i)
    elif i == lista_archivos[4]:
        genero = archivo_a_lista(i)
    elif i == lista_archivos[5]:
        promedio = archivo_a_lista(i)
    elif i == lista_archivos[6]:
        region = archivo_a_lista(i)

for i in range(len(edad)):
    edad[i] = int(edad[i])
for i in range(len(promedio)):
    promedio[i] = float(promedio[i])

estudiantes = []
for i in range(len(edad)):
    estudiantes.append(i)

dicc_lab1 = {}
for i in estudiantes:
    dicc_lab1[i]=[edad[i],escolaridad[i],estado_civil[i],estrato[i],genero[i],promedio[i],region[i]]
```

```python
# código
#LAURA I. PINILLA MESA

lista_estrato_1 = []
lista_estrato_2 = []
lista_estrato_3 = []
lista_estrato_4 = []
lista_estrato_5 = []
lista_estrato_6 = []

for i in range(len(estudiantes)):
    if dicc_lab1[i][3] == "1":
       valor_1 = [i, dicc_lab1[i][5]]
       lista_estrato_1.append(valor_1)
    elif dicc_lab1[i][3] == "2":
       valor_2 = [i, dicc_lab1[i][5]]
       lista_estrato_2.append(valor_2)
    elif dicc_lab1[i][3] == "3":
       valor_3 = [i, dicc_lab1[i][5]]
       lista_estrato_3.append(valor_3)
    elif dicc_lab1[i][3] == "4":
       valor_4 = [i, dicc_lab1[i][5]]
       lista_estrato_4.append(valor_4)
    elif dicc_lab1[i][3] == "5":
       valor_5 = [i, dicc_lab1[i][5]]
       lista_estrato_5.append(valor_5)

lista_estrato_1.sort(key=lambda x: x[1])
lista_estrato_1.reverse()
lista_estrato_2.sort(key=lambda x: x[1])
lista_estrato_2.reverse()
lista_estrato_3.sort(key=lambda x: x[1])
lista_estrato_3.reverse()
lista_estrato_4.sort(key=lambda x: x[1])
lista_estrato_4.reverse()
lista_estrato_5.sort(key=lambda x: x[1])
lista_estrato_5.reverse()

def asignar(lista_asignar):
    numero_asignar = len(lista_asignar)*0.02
    numero_red = math_.floor(numero_asignar)
    grupo_asignar = []
    contador = 0
    for i in range(len(lista_asignar)):
        id_ = 0
        if contador < numero_red:
            contador = contador+1
            id_ = lista_asignar[i][0]
            grupo_asignar.append(id_)
    return grupo_asignar

def sacar_becados(lista_original,lista_becados):
    lista_sin_becados = []
    for i in range(len(lista_original)):
        id_2 = lista_original[i][0]
        if id_2 not in lista_becados:
            lista_sin_becados.append(lista_original[i])
    return lista_sin_becados

def recorrer_estratos(becas_disp,lista_e1,lista_e2,lista_e3,lista_e4,lista_e5):
    estudiantes_asignar = []
    #estrato1
    if becas_disp > 0:
        asignar_e1 = asignar(lista_e1)
        if becas_disp >= len(asignar_e1):
            num_asignar_e1 = len(asignar_e1)
        else:
            num_asignar_e1 = becas_disp
        becas_disp = becas_disp - num_asignar_e1
        estudiantes_e1 = asignar_e1[:num_asignar_e1]
        estudiantes_asignar.extend(estudiantes_e1)
        #estrato2
        if becas_disp > 0:
            asignar_e2 = asignar(lista_e2)
            if becas_disp >= len(asignar_e2):
                num_asignar_e2 = len(asignar_e2)
            else:
                num_asignar_e2 = becas_disp
            becas_disp = becas_disp - num_asignar_e2
            estudiantes_e2 = asignar_e2[:num_asignar_e2]
            estudiantes_asignar.extend(estudiantes_e2)
            #estrato3
            if becas_disp > 0:
                asignar_e3 = asignar(lista_e3)
                if becas_disp >= len(asignar_e3):
                    num_asignar_e3 = len(asignar_e3)
                else:
                    num_asignar_e3 = becas_disp
                becas_disp = becas_disp - num_asignar_e3
                estudiantes_e3 = asignar_e3[:num_asignar_e3]
                estudiantes_asignar.extend(estudiantes_e3)
                #estrato4
                if becas_disp > 0:
                    asignar_e4 = asignar(lista_e4)
                    if becas_disp >= len(asignar_e4):
                        num_asignar_e4 = len(asignar_e4)
                    else:
                        num_asignar_e4 = becas_disp
                    becas_disp = becas_disp - num_asignar_e4
                    estudiantes_e4 = asignar_e4[:num_asignar_e4]
                    estudiantes_asignar.extend(estudiantes_e4)
                    #estrato5
                    if becas_disp > 0:
                        asignar_e5 = asignar(lista_e5)
                        if becas_disp >= len(asignar_e5):
                            num_asignar_e5 = len(asignar_e5)
                        else:
                            num_asignar_e5 = becas_disp
                        becas_disp = becas_disp - num_asignar_e5
                        estudiantes_e5 = asignar_e5[:num_asignar_e5]
                        estudiantes_asignar.extend(estudiantes_e5)
            
    return estudiantes_asignar

def puedo_asignar(lista_e1,lista_e2,lista_e3,lista_e4,lista_e5):
    puedo = 0
    p_e1 = len(lista_e1)*0.02
    p_e1_red = math_.floor(p_e1)
    p_e2 = len(lista_e2)*0.02
    p_e2_red = math_.floor(p_e2)
    p_e3 = len(lista_e3)*0.02
    p_e3_red = math_.floor(p_e3)
    p_e4 = len(lista_e4)*0.02
    p_e4_red = math_.floor(p_e4)
    p_e5 = len(lista_e5)*0.02
    p_e5_red = math_.floor(p_e5)
    if p_e1_red >= 1 or p_e2_red >= 1 or p_e3_red >= 1 or p_e4_red >= 1 or p_e5_red >= 1:   
        puedo = 1
    return puedo

def mision4(numbecas,lista_e1,lista_e2,lista_e3,lista_e4,lista_e5):
    
    lista_becados_final = []
    becas = numbecas
    lista_filtrada_1 = lista_e1
    lista_filtrada_2 = lista_e2
    lista_filtrada_3 = lista_e3
    lista_filtrada_4 = lista_e4
    lista_filtrada_5 = lista_e5

    while becas > 0:
        puedo_asignar_m4 = puedo_asignar(lista_filtrada_1, lista_filtrada_2, lista_filtrada_3, lista_filtrada_4, lista_filtrada_5)
        if puedo_asignar_m4 == 1:
            recorrido = recorrer_estratos(becas, lista_filtrada_1, lista_filtrada_2, lista_filtrada_3, lista_filtrada_4, lista_filtrada_5)
            lista_becados_final.extend(recorrido)
            becas = becas - len(recorrido)
            lista_filtrada_1 = sacar_becados(lista_estrato_1, lista_becados_final)
            lista_filtrada_2 = sacar_becados(lista_estrato_2, lista_becados_final)
            lista_filtrada_3 = sacar_becados(lista_estrato_3, lista_becados_final)
            lista_filtrada_4 = sacar_becados(lista_estrato_4, lista_becados_final)
            lista_filtrada_5 = sacar_becados(lista_estrato_5, lista_becados_final)
        else:
            break
    
    return lista_becados_final

#A modo de ejemplo, se toma un número de 40 becas disponibles para mostrar los estudiantes becados con este método
print(mision4(40,lista_estrato_1,lista_estrato_2,lista_estrato_3,lista_estrato_4,lista_estrato_5))      
```

```python
# código
#LAURA I. PINILLA MESA
    
def filtrar_edad(base_completa,edad_i,edad_s):
    filtrado_edad=[]
    for i in range(len(base_completa)):
        if base_completa[i][0] >= edad_i and base_completa[i][0] <= edad_s:
            id_3 = i
            filtrado_edad.append(id_3)  
    return filtrado_edad

def becas_x_grupo(num_becas,dupla_prop):
    becas_grupo = {}
    for i in range(len(dupla_prop)):
        categoria = dupla_prop[i][0]
        proporcion = dupla_prop[i][1]
        calc_becas = math_.floor(num_becas*proporcion)
        #becas_cat = [categoria,calc_becas]
        becas_grupo[categoria] = calc_becas
    return becas_grupo

def asignar3(lista_grupo,becas_disp_grupo,filtro_edad):
    total_asignar_grupo = 0
    tam_grupo = len(lista_grupo)
    if becas_disp_grupo > tam_grupo:
        total_asignar_grupo = tam_grupo
    else:
        total_asignar_grupo = becas_disp_grupo
        
    grupo_asignar = []
    contador = 0
    for i in range(len(lista_grupo)):
        id_4 = 0
        if contador < total_asignar_grupo:
            id_4 = lista_grupo[i][0]
            if id_4 in filtro_edad:
                contador = contador+1
                grupo_asignar.append(id_4)
    
    return grupo_asignar

def mision7(base_original,numbecas,edad_inf,edad_sup,criterio,proporciones):
    
    lista_becados_final_3 = []
    lista_1 = []
    lista_2 = []
    lista_3 = []
    lista_4 = []
    lista_5 = []
    
    #armar grupos
    if criterio == "genero":
        for i in range(len(base_original)):
            if base_original[i][4] == "femenino":
                valor_1 = [i, base_original[i][5]]
                lista_1.append(valor_1)
            elif base_original[i][4] == "masculino":
                valor_2 = [i, base_original[i][5]]
                lista_2.append(valor_2)
            elif base_original[i][4] == "otro":
                valor_3 = [i, base_original[i][5]]
                lista_3.append(valor_3)
            elif base_original[i][4] == "no binario":
                valor_4 = [i, base_original[i][5]]
                lista_4.append(valor_4)
        
    elif criterio == "estrato":
        for i in range(len(base_original)):
            if base_original[i][3] == "1":
                valor_1 = [i, base_original[i][5]]
                lista_1.append(valor_1)
            elif base_original[i][3] == "2":
                valor_2 = [i, base_original[i][5]]
                lista_2.append(valor_2)
            elif base_original[i][3] == "3":
                valor_3 = [i, base_original[i][5]]
                lista_3.append(valor_3)
            elif base_original[i][3] == "4":
                valor_4 = [i, base_original[i][5]]
                lista_4.append(valor_4)
            elif base_original[i][3] == "5":
                valor_5 = [i, base_original[i][5]]
                lista_5.append(valor_5)
    
    elif criterio == "region":
        for i in range(len(base_original)):
            if base_original[i][6] == "Andina":
                valor_1 = [i, base_original[i][5]]
                lista_1.append(valor_1)
            elif base_original[i][6] == "Pacifica":
                valor_2 = [i, base_original[i][5]]
                lista_2.append(valor_2)
            elif base_original[i][6] == "Orinoquia":
                valor_3 = [i, base_original[i][5]]
                lista_3.append(valor_3)
            elif base_original[i][6] == "Caribe":
                valor_4 = [i, base_original[i][5]]
                lista_4.append(valor_4)
            elif base_original[i][6] == "Amazonia":
                valor_5 = [i, base_original[i][5]]
                lista_5.append(valor_5)
    
    listas_grupos = []
    if criterio == "genero":
        listas_grupos = [lista_1,lista_2,lista_3,lista_4]
    else:
        listas_grupos = [lista_1,lista_2,lista_3,lista_4,lista_5]
    
    for i in listas_grupos:
        i.sort(key=lambda x: x[1])
        i.reverse()
    
    filtro_edades = filtrar_edad(base_original,edad_inf,edad_sup)
    becas_disponibles_grupos = becas_x_grupo(numbecas,proporciones)
    
    for i in listas_grupos:
        if criterio == "genero":
            if i == lista_1:
                nombre_lista = "femenino"
            elif i == lista_2:
                nombre_lista = "masculino"
            elif i == lista_3:
                nombre_lista = "otro"
            else:
                nombre_lista = "no binario"
        
        elif criterio == "estrato":
            if i == lista_1:
                nombre_lista = "1"
            elif i == lista_2:
                nombre_lista = "2"
            elif i == lista_3:
                nombre_lista = "3"
            elif i == lista_4:
                nombre_lista = "4"
            else:
                nombre_lista = "5"    
        
        elif criterio == "region":
            if i == lista_1:
                nombre_lista = "Andina"
            elif i == lista_2:
                nombre_lista = "Pacifica"
            elif i == lista_3:
                nombre_lista = "Orinoquia"
            elif i == lista_4:
                nombre_lista = "Caribe"
            else:
                nombre_lista = "Amazonia"
        
        for j in becas_disponibles_grupos:
            cat = j
            if cat == nombre_lista:
                categ = cat
        
        becas_disponible_grupo = becas_disponibles_grupos[categ]
        asignados_m7 = asignar3(i,becas_disponible_grupo,filtro_edades)
        lista_becados_final_3.extend(asignados_m7)
    
    return lista_becados_final_3
```
